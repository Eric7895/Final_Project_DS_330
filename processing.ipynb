```python
import pandas as pd
import numpy as np
```

```python
# Read csv file
data = pd.read_csv("sleep.csv")
data.head()
```

```python
# split string blood pressure to numeric columns
data[["Systolic", "Diastolic"]] = data["Blood Pressure"].str.split("/", expand=True)
data.head()
```

```python
age = data["Age"]
duration = data["Sleep Duration"]
quailty = data["Quality of Sleep"]
physical = data["Physical Activity Level"]
stress = data["Stress Level"]
heart = data["Heart Rate"]
daily = data["Daily Steps"]
systolic = int(data["Systolic"])
diastolic = int(data["Diastolic"])
```

```python
age_class = age.unique()
duration_class = duration.unique()
```

```python
def frequence_count(class_label, data):
    count = [0 for i in range(len(class_label))]
    for i in range(len(data)):
        for j in range(len(class_label)):
            if data[i] == class_label[j]:
                count[j] += 1
    return np.array(count)
```

```python
age_count = frequence_count(age_class, age)
dur_count = frequence_count(duration_class, duration)
```

```python
result = pd.DataFrame({"name": duration_class, "number": dur_count}, columns=["name", "number"])
result.head()
```

```python
result.to_csv('duration.csv', index=False)
```
